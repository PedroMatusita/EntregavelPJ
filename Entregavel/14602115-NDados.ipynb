## Questão 2
a) Crie uma coluna que corresponda a soma do preço e da taxa de serviço (price + service fee)

```python
SOMA = []

for index in range(df.shape[0]):
    PRECO = df.loc[index,'price']
    SERVICE = df.loc[index,'service fee']
    
#  tirar cifrao e virgulas
    PRECO = PRECO.replace('$','')
    PRECO = PRECO.replace(',','')
    PRECO = int(PRECO)
    
    if isinstance(SERVICE, str):
        SERVICE = SERVICE.replace("$",'')
        SERVICE = SERVICE.replace(",",'')
        SERVICE = int(SERVICE)

        SOMA.append(PRECO+SERVICE)

    else:
#       preco sem taxa de servico
        SOMA.append(PRECO)
    
# colocar cifrao
# for i in range(len(SOMA)):
#     SOMA[i] = '$' + str(SOMA[i])
    
df['price + service fee'] = SOMA

df_subset = df.loc[:, ['price', 'service fee', 'price + service fee']]
print(df_subset)
```

b) Obtenha o nome do imóvel que possui o maior número de avaliações (number of reviews) e que apresente taxa média de avaliação igual a 5 (review rate number).

```python
#Estratégia: varrer o dataframe checando avaliação e n° de avaliações

nREVIEW = 0
NAMEBEST = None

for index, row in df.iterrows():
    if row['number of reviews'] > nREVIEW and row['review rate number'] == 5.0:
        nREVIEW = row['number of reviews']
        NAMEBEST = row['NAME']

print(f"Melhor imóvel: {NAMEBEST}.")
```

```python
# Estratégia: Usar um dicionario com chave == nome do bairro e valor numero de vezes (percorrer uma vez so)
# percorrer outra vez e fazer a media

BAIRROS = {}

for index, row in df.iterrows():
    if row['neighbourhood'] in BAIRROS:
        BAIRROS[row['neighbourhood']] += 1
    else:
        BAIRROS[row['neighbourhood']] = 1

        
# Descobrir maior
MAIOR = 0
mBAIRRO = None

for chave in BAIRROS:
    if BAIRROS[chave] > MAIOR:
        mBAIRRO = chave
        MAIOR = BAIRROS[chave]
        
print(f"Bairro que mais aparece: {mBAIRRO}.")

# Média
contagem = 0
soma = 0

for index, row in df.iterrows():
    if row['neighbourhood'] == mBAIRRO:
        contagem += 1
        
        PRECO = row['price + service fee']
        soma += PRECO
 
print(f"Média do preço do bairro: {soma/contagem}.")
```

a) Crie um gráfico de latitude (lat) e longitude (long) dividido por distritos (neighbourhood group), o qual retornará o mapa de Nova York.

## Etapa 2 - Estimativa de valor de mercado com Regressão
### Desafio Opcional

Com os clientes já bem alinhados em relação às particularidades do mercado imobiliário da cidade, chegou a hora de avaliar as oportunidades de investimento em Araraquara. Para tal, o cliente pediu uma análise de regressão como forma de estimar o valor de mercado real das listagens a partir das vendas realizadas no passado na cidade, e comparar com os preços pedidos nas listagens.

De forma simples, um algoritmo de regressão é um modelo de Machine Learning capaz de prever um valor associado à um conjunto de variáveis. Isto é, a partir de uma base de treinamento com valores finais já fornecidos, o modelo aprende a indentificar características menos ou mais valiosas para o comportamento deste valor final, e utiliza isso para estimar um valor final para uma base que ainda não possui o valor definido. 

Um exemplo simples pode ser o salário oferecido à jogadores de futebol:
- sabe-se que o jogador 1 teve 2 gols, 4 assistências, 72% de aproveitamento e recebeu 100 mil reais
- sabe-se que o jogador 2 teve 5 gols, 2 assistências, 64% de aproveitamento e recebeu 85 mil reais
- sabe-se que o jogador 3 teve 4 gols, 2 assistências, 67% de aproveitamento, quanto seria um salário justo para esse jogador?

Os modelos de regressão prometem fornecer uma resposta adequada para esta pergunta, levando em consideração tudo aquilo que foi analisado em sua base de teste, para prever um valor correspondente adequado.

A ideia da Poli Estates é utilizar a base de históricos de venda de imóveis da cidade para treinar o modelo, e utilizar toda a inteligência adquirida nesta etapa para identificar o valor de mercado das listagens atuais disponíveis no mercado. Se a listagem possuir um valor de mercado inferior ao pedido pelo proprietário, esta listagem pode ser considerada um mau investimento, porém, se o proprietário indicar um valor de venda inferior ao valor de mercado aferido pela análise de regressão, este pode ser considerado um bom investimento para a companhia.

Recomenda-se a utilização das diferentes opções de modelos de regressão implementados na biblioteca sklearn, e que sejam testados para descobrir a melhor performance para a base de dados estudada (recomenda-se o teste por validação cruzada). É importante ressaltar a necessidade de filtrar as características abordadas pelo modelo, de forma a não sobrecarregar o algoritmo com features (colunas utilizadas como varáveis independentes) dos imóveis que sejam pouco correlatas, ou capazes de gerar ruídos nas análises. Após o desenvolvimento e melhoria do modelo de regressão na base do histórico, aplicar o resultado sobre a base de vendas e identificar as melhores oportunidades de investimento para a companhia. Apresentar um relatório com o veredito sobre cada uma das propriedades ofertadas no mercado.

Para trazer esse veredito, basta aplicar a função predict do modelo que for treinado aos dados da base de vendas comentada. ## Modelo 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# Substitua 'target_column' pelo nome da coluna que você deseja prever
target_column = 'price'

# Seleciona as características (variáveis independentes)
features = hist.drop(target_column, axis=1)

# Seleciona a variável alvo (variável dependente)
target = hist[target_column]

# Divide os dados em conjunto de treinamento e conjunto de teste
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=9)

# Cria o modelo de regressão linear
modelo = LinearRegression()

# Treina o modelo com os dados de treinamento
modelo.fit(X_train, y_train)

# Realiza previsões no conjunto de teste
y_pred = modelo.predict(X_test)

# Avalia o desempenho do modelo
r2 = r2_score(y_test, y_pred)
print(f'R^2 Score: {r2}')

# Visualiza as previsões vs. valores reais (opcional)
plt.scatter(y_test, y_pred)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Previsões vs. Valores Reais')
plt.show()
```

Com R^2 score de aproximadamente 0.88 o modelo é eficaz, com o gráfico mostrando quase uma linha reta
